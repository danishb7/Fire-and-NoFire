```python
import os
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet50
from PIL import Image
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

```

```python
# Paths
data_dir = "../FiSmo-Images/Flickr-Fire/Flickr-Fire_flame"
flames_file = "../FiSmo-Images/Flickr-Fire/flamesId.txt"
not_flames_file = "../FiSmo-Images/Flickr-Fire/notFlamesId.txt"
model_path = "deepfire_resnet50.pth"  # Update with your model path

# Load image paths from .txt files
def load_image_list(file_path, label, data_dir):
    with open(file_path, "r") as f:
        images = [os.path.join(data_dir, line.strip()) for line in f.readlines()]
    return [(img, label) for img in images]

# Load dataset
fire_images = load_image_list(flames_file, 1, data_dir)
no_fire_images = load_image_list(not_flames_file, 0, data_dir)
dataset = fire_images + no_fire_images

# Transform for ResNet50
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

```

```python
def count_images(directory):
    supported_formats = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif')  # Supported image formats
    total_images = 0

    for root, _, files in os.walk(directory):
        total_images += sum(1 for file in files if file.lower().endswith(supported_formats))
    
    return total_images

# Get total image count
data_dir = "../FiSmo-Images/Flickr-Fire/Flickr-Fire_flame"
total_images = count_images(data_dir)
print(f"Total images in '{data_dir}': {total_images}")

# data_dir = "../FiSmo-Images/Flickr-FireSmoke/imgs"
# total_images = count_images(data_dir)
# print(f"Total images in '{data_dir}': {total_images}")
```

```python
# Load model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = resnet50()
model.fc = torch.nn.Linear(model.fc.in_features, 2)  # Assuming binary classification
model.load_state_dict(torch.load(model_path, map_location=device))
model.to(device)
model.eval()

```

```python
# Test loop
# err_count = 0
def test_model(dataset, model, transform, device):
    y_true = []
    y_pred = []

    for img_path, label in dataset:
        try:
            image = Image.open(img_path).convert("RGB")
            input_tensor = transform(image).unsqueeze(0).to(device)

            # Predict
            with torch.no_grad():
                output = model(input_tensor)
                pred_label = torch.argmax(output, dim=1).item()

            y_true.append(label)
            y_pred.append(pred_label)

        except Exception as e:
            print(f"Error loading image {img_path}: {e}")

    return y_true, y_pred

# Run the test
y_true, y_pred = test_model(dataset, model, transform, device)

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label=1)
recall = recall_score(y_true, y_pred, pos_label=1)
f1 = f1_score(y_true, y_pred, pos_label=1)

# Display metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

print(err_count)
```
